```python
def run_quiz(questions, timed=True):
    """Run an interactive multiple-choice quiz."""
    random.shuffle(questions)
    
    score = 0
    total = len(questions)
    results_by_category = {}
    wrong_questions = []
    
    start_time = time.time()
    
    print("=" * 60)
    print(f"  RAPID FIRE QUIZ — {total} Questions")
    if timed:
        print("  Timer is running! Target: 15 minutes")
    print("=" * 60)
    print("  Type A, B, C, or D and press Enter.")
    print("  Type 'q' to quit early.")
    print("=" * 60 + "\n")
    
    for i, q in enumerate(questions, 1):
        cat = q["category"]
        if cat not in results_by_category:
            results_by_category[cat] = {"correct": 0, "total": 0}
        results_by_category[cat]["total"] += 1
        
        elapsed = time.time() - start_time
        mins, secs = divmod(int(elapsed), 60)
        
        print(f"--- Question {i}/{total} [{cat}] [{mins:02d}:{secs:02d}] ---")
        print(f"\n{q['question']}\n")
        for key, val in q["options"].items():
            print(f"  {key}) {val}")
        
        while True:
            ans = input("\nYour answer: ").strip().upper()
            if ans in ('A', 'B', 'C', 'D', 'Q'):
                break
            print("Invalid input. Enter A, B, C, D, or Q.")
        
        if ans == 'Q':
            print("\nQuiz ended early.")
            total = i - 1
            results_by_category[cat]["total"] -= 1
            break
        
        if ans == q["answer"]:
            print("\n✅ Correct!")
            score += 1
            results_by_category[cat]["correct"] += 1
        else:
            print(f"\n❌ Wrong! The answer is {q['answer']}) {q['options'][q['answer']]}")
            wrong_questions.append(q)
        
        print(f"   💡 {q['explanation']}\n")
    
    # ---- RESULTS ----
    elapsed = time.time() - start_time
    mins, secs = divmod(int(elapsed), 60)
    
    print("\n" + "=" * 60)
    print("  QUIZ RESULTS")
    print("=" * 60)
    
    if total > 0:
        pct = score / total * 100
        print(f"\n  Score: {score}/{total} ({pct:.0f}%)")
        print(f"  Time:  {mins:02d}:{secs:02d}")
        
        print(f"\n  {'Category':<30} {'Score':>10}  {'Pct':>6}")
        print("  " + "-" * 50)
        for cat, res in sorted(results_by_category.items()):
            if res['total'] > 0:
                cat_pct = res['correct'] / res['total'] * 100
                bar = "🟢" if cat_pct >= 80 else "🟡" if cat_pct >= 60 else "🔴"
                print(f"  {bar} {cat:<28} {res['correct']:>3}/{res['total']:<3}   {cat_pct:>5.0f}%")
        
        if wrong_questions:
            print(f"\n  ⚠️  Review these {len(wrong_questions)} topics:")
            for q in wrong_questions:
                print(f"     - [{q['category']}] {q['question']}")
                print(f"       Answer: {q['answer']}) {q['options'][q['answer']]}")
        
        if pct >= 90:
            print("\n  🎯 Excellent! You're well prepared.")
        elif pct >= 70:
            print("\n  👍 Good, but review the topics above.")
        else:
            print("\n  📚 Keep studying — focus on the red categories.")
    
    print("\n" + "=" * 60)
```

```python
# 🚀 RUN THE QUIZ!
# Re-run this cell anytime to take the quiz again (questions will be reshuffled)
run_quiz(questions)
```
